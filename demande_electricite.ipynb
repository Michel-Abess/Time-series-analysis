### Checking stationarity

#### Adfuller test

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
result_test = adfuller(x_conso_cte_2016.values)
print('ADF Statistic: %f' % result_test[0])
print('p-value: %f' %result_test[1])
```

p-value < 5% we can reject H0 and admit stationarity of the series

```python
# Other method
from statsmodels.tsa.stattools import adfuller
print('Result of Dickey-Fuller test:')
dftest = adfuller(x_conso_cte_2016['conso_cte'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of observations uUsed'])
for key, value in dftest[4].items():
    dfoutput['critical value (%s)'%key] = value
    
print(dfoutput)
```

the augmented Dickey fuller test seems to show that the serie is stationary
H0 hyoothesis => the serie is non stationary
H1 hyoothesis => the serie is stationary

p-value < 5% so we can adopt H1 hypothesis knowing that the H0 is less likely to occur. In addition of that Test statistic is less than the critical value for 1%. Hence we can reject the null hypothesis

The ACF output of the series shows a slow decay towards 0. Also, the autocorrelation is positive for 10+ lags, this autocorrelogram is therefore not simple and needs further differencing. We perform a differentiation X(t) - X(t-1) to solve the non-stationarity problem

#### Modèle 1

First, we estimate a model $SARIMA(1,1,1)(1,1,1)_{12}$ on the base of simple and partial empirical autocorrelograms.

```python
from statsmodels.tsa.statespace.sarimax import *
from statsmodels.stats.diagnostic import acorr_ljungbox

model1 = SARIMAX(np.asarray(x['conso_cte']), order=(1,1,1), seasonal_order=(1,1,1,12))
results1 = model1.fit()
print(results1.summary())

print('Retard : p-value')
for elt in [6, 12, 18, 24, 30, 36]:
    print('{} : {}'.format(elt, acorr_ljungbox(results1.resid, lags=elt).mean()))
```

one parameter of this model seems to be insignificant. we have a p-value of 16% for the coefficient of the seasonal AR model. The parameter AR.S.L12 is not significantly different from 0.
    
The Jung box test is conclusive: Residues follow a white noise



#### Modèle 2


we remove the insignificant parameter estimated in the first model and we obtain the second model


```python
model2 = SARIMAX(np.asarray(x['conso_cte']), order=(1,1,1), seasonal_order=(0,1,1,12))
results2 = model2.fit()
print(results2.summary())

print('Retard : p-value')
for elt in [6, 12, 18, 24, 30, 36]:
    print('{} : {}'.format(elt, acorr_ljungbox(results2.resid, lags=elt).mean()))
```

All the parameters of the model have p-value lower than 5%. But zero is still in the confidence interval for two parameters ar.L1  and ma.S.L12, so there is a chance that they get zero at value
Also the whiteness of the residuals is validated at the 5% test level

```python
from scipy.stats import shapiro

shapiro(results2.resid)
```

The normality test of the residuals is also validated for this model.
